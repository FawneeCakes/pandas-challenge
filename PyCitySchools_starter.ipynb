# **PyCity Schools Analysis**

1.   Budget vs. Test Results:

*   Overall, schools with higher budgets did not achieve better test results. Surprisingly, schools spending between $645 and $675 per student actually performed worse than those with smaller budgets (around $585 per student).

2.   School Size and Math Performance:

*   Smaller and medium-sized schools significantly outperformed large schools in terms of math performance. The passing rates for math were 89-91% in smaller and medium-sized schools, compared to only 67% in large schools.

3.   Charter Schools vs. Public District Schools:

*   Charter schools, as a whole, outperformed public district schools across all metrics. However, further analysis is needed to determine whether this effect is due to school practices or simply because charter schools tend to serve smaller student populations.












```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)


# Combine the data into a single dataset.
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# Use the provided code to calculate the passing rates
# Calculate the percentage of students who passed math per school
school_passing_math = school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name")["math_score"].count()
per_school_passing_math = school_passing_math / per_school_counts * 100

# Calculate the percentage of students who passed reading per school
school_passing_reading = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name")["reading_score"].count()
per_school_passing_reading = school_passing_reading / per_school_counts * 100

# Calculate the percentage of students who passed both math and reading per school
school_passing_math_and_reading = school_data_complete[(school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)].groupby("school_name")["student_name"].count()
per_school_passing_math_and_reading = school_passing_math_and_reading / per_school_counts * 100

# Print the passing rates per school
print("Passing math percentage per school:\n", per_school_passing_math)
print("\nPassing reading percentage per school:\n", per_school_passing_reading)
print("\nOverall passing percentage per school:\n", per_school_passing_math_and_reading)
```
